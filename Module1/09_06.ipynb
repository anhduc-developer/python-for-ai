```python
import math
def calc_f1_score(tp, fp, fn):
    precision = tp / (tp + fp)
    recall = tp / (tp + fn)
    f1_score = 2 * (precision * recall) / (precision + recall)
    return f1_score
assert round(calc_f1_score(tp = 2, fp = 3, fn = 5), 2) == 0.33
print(round(calc_f1_score(tp = 2, fp = 3, fn = 5), 2))

```

```python
import math
def is_number(n):
    try:
        float(n)
    except ValueError:
        return False
    return True

assert is_number(3) == 1.0
assert is_number('-2a') == 0.0
print(is_number(1))
print(is_number('n'))
```

```python
def calc_activation_function(x, act_name):
    def calc_relu(x):
        if x <= 0:
            result = 0.0
        else:
            result = x
        return float(result)
    def calc_elu(x):
        alpha = 0.01
        result = None
        if x < 0:
            result = alpha * (math.e ** x - 1)
        else:
            result = x
        return result
    def calc_sig(x):
        return 1 / (1 + math.e ** (-x))
    result = None
    if act_name == "sigmoid":
        result =  calc_sig(x)
    elif act_name == "elu":
        result =  calc_elu(x)
    elif act_name == "relu":
        result = calc_relu(x)
    return result
assert round(calc_activation_function(x = 1, act_name='relu')) == 1
```

```python
def exercise_2():
    x = input('Input x = ')
    if not is_number(x):
        print('x must be a number')
        return
    act_name = input('Input activation Function (sigmoid | relu | elu): ')
    x = float(x)
    result = calc_activation_function(x, act_name)
    if result is None:
        print(f'{act_name} is not supported')
    else:
        print(f'{act_name} = {result}')
exercise_2()
```

```python
def is_numeric(n):
    try:
        int(n)
    except:
        return False
    return True
```

```python
import random
def calc_MAE(n):
    mean = 0
    for i in range(n):
        y = random.random()
        y_hat = random.random()
        abs_error = abs(y - y_hat)
        mean += abs_error
    mean /= n
    return mean
print(calc_MAE(3))
```

```python
import random
def calc_MSE(n):
    mean = 0
    for i in range(n):
        y = random.random()
        y_hat = random.random()
        abs_error = (y - y_hat) ** 2
        mean += abs_error
    return mean
print(calc_MSE(3))
```

```python
import random
def calc_RMSE(n):
    return math.sqrt(calc_MSE(n))
print(calc_RMSE(3))
```

```python
def calc_regress_loss_function(n, regresse_loss_name):
    loss = None
    if regresse_loss_name == "MAE":
        loss =  calc_MAE(n)
    elif regresse_loss_name == "MSE":
        loss =  calc_MSE(n)
    elif regresse_loss_name == "RMSE":
        loss = calc_RMSE(n)
    return loss
```

```python
def exercise_3():
    n = input("Input n = ")
    if not is_numeric(n):
        print('n must be a number')
        return
    regress_loss_name = input('Input regress loss function (mae or mse or remse)')
    n = int(n)
    result = calc_regress_loss_function(n, regress_loss_name)
    if result is None:
        print(f'{regress_loss_name} is not supported')
    else:
        print(f'{regress_loss_name} = {result}')
exercise_3()
```

```python
def check_armstrong_num(n):
    if n < 18: return False
    init = n
    result = 0
    while n > 0:
        temp = n % 10
        result += temp ** 3
        n //= 10
    return result == init
print(check_armstrong_num(153))
def compute_list_armstrong(data):
    result = []
    for x in data:
        if check_armstrong_num(x):
            result.append(x)
    return result


if __name__ == "__main__":
    arr = list(map(int, input().split()))
    print(compute_list_armstrong(arr))
```

```python
def compute_list_candy(arr, extra_candies):
    max_candy = max(arr)
    result = []
    for x in arr:
        if x + extra_candies >= max_candy:
            result.append(True)
        else:
            result.append(False)
    return result


if __name__ == "__main__":
    arr = list(map(int, input().split()))
    extra_candies = int(input())
    print(compute_list_candy(arr, extra_candies))
```

```python
def compute_median(arr):
    n = len(arr)
    arr.sort()
    if n % 2 == 1:
        index = (n + 1) // 2 - 1
        return arr[index]
    else:
        index1 = n // 2 - 1
        index2 = n // 2
        return (arr[index1] + arr[index2]) / 2


arr = list(map(int, input().split()))
print(compute_median(arr))
```
