```python
import math
def calc_f1_score(tp, fp, fn):
    precision = tp / (tp + fp)
    recall = tp / (tp + fn)
    f1_score = 2 * (precision * recall) / (precision + recall)
    return f1_score
assert round(calc_f1_score(tp = 2, fp = 3, fn = 5), 2) == 0.33
print(round(calc_f1_score(tp = 2, fp = 3, fn = 5), 2))

```

```python
import math
def is_number(n):
    try:
        float(n)
    except ValueError:
        return False
    return True

assert is_number(3) == 1.0
assert is_number('-2a') == 0.0
print(is_number(1))
print(is_number('n'))
```

```python
def calc_activation_function(x, act_name):
    def calc_relu(x):
        if x <= 0:
            result = 0.0
        else:
            result = x
        return float(result)
    def calc_elu(x):
        alpha = 0.01
        result = None
        if x < 0:
            result = alpha * (math.e ** x - 1)
        else:
            result = x
        return result
    def calc_sig(x):
        return 1 / (1 + math.e ** (-x))
    result = None
    if act_name == "sigmoid":
        result =  calc_sig(x)
    elif act_name == "elu":
        result =  calc_elu(x)
    elif act_name == "relu":
        result = calc_relu(x)
    return result
assert round(calc_activation_function(x = 1, act_name='relu')) == 1
```

```python
def exercise_2():
    x = input('Input x = ')
    if not is_number(x):
        print('x must be a number')
        return
    act_name = input('Input activation Function (sigmoid | relu | elu): ')
    x = float(x)
    result = calc_activation_function(x, act_name)
    if result is None:
        print(f'{act_name} is not supported')
    else:
        print(f'{act_name} = {result}')
exercise_2()
```
