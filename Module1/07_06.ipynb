```python
# PI Estimation
def compute_PI_Gregory(n):
  result = 0
  for i in range(1, n + 1):
    result += ((-1) ** (i + 1)) / (2 * i - 1)
  return result * 4

def compute_PI_Nilakantha(n):
  result = 3
  for i in range(n + 1):
    result += 4 * (((-1) ** i) / ((2 * i + 2) * (2 * i + 3) * (2 * i + 4)))
  return result

#Newton Method
def compute_quadratic(a, n):
  x = a / 2
  for i in range(n):
    x = (x + a / x) / 2
  return x


print(compute_PI_Gregory(1000))
print(compute_PI_Nilakantha(1000))
print(compute_quadratic(9, 3))
```
