```python
import random
import math
# PI Estimation
def compute_PI_Gregory(n):
  result = 0
  for i in range(1, n + 1):
    result += ((-1) ** (i + 1)) / (2 * i - 1)
  return result * 4

def compute_PI_Nilakantha(n):
  result = 3
  for i in range(n + 1):
    result += 4 * (((-1) ** i) / ((2 * i + 2) * (2 * i + 3) * (2 * i + 4)))
  return result

#Newton Method
def compute_quadratic(a, n):
  x = a / 2
  for i in range(n):
    x = (x + a / x) / 2
  return x
def random_function():
  while True:
    num = random.randint(0, 10)
    if num == 5:
      print("Hello World!")
      break

#Compute the area of a unut circle
def compute_y(x):
  return math.sqrt(1 - x ** 2)
def compute_half_area(lim, delta_x):
  n = int(lim / delta_x)
  half_area = 0
  x = -1.0
  for _ in range(n):
    y = compute_y(x)
    half_area += y * delta_x
    x = x + delta_x
  return half_area

print(compute_PI_Gregory(1000))
print(compute_PI_Nilakantha(1000))
print(compute_quadratic(9, 3))
print(compute_half_area(2, 0.1))
random_function()
```
