```python
#Evaluate classification by f1-score

def compute_f1_score(tp, fp, fn):
    #compute precision
    precision = tp / (tp + fp)
    print(f"precision = {precision:.2f}")
    #compute recall
    recall = tp / (tp + fn)
    print(f"recall = {recall:.2f}")
    #compute_f1_score
    f1_score = (2 * precision * recall) / (precision + recall)
    return f1_score
def compute_Precision(tp, fp):
    return tp / (tp + fp)
def compute_Recall(tp, fn):
    return tp / (tp + fn)
def compute_PI(n):
    result = 0
    for i in range(1, n + 1):
        result += 4 * (((-1) ** (i + 1)) / (2 * i - 1))
    return result
def compute_PI_v2(n):
    result = 3
    for i in range(n + 1):
        result += 4 * (((-1) ** i) / ((2 * i + 2) * (2 * i + 3) * (2 * i +4)))
    return result
if __name__ == "__main__":
    # tp, fp, fn = map(int ,input().split())
    # print(compute_f1_score(tp, fp, fn))
    print(compute_PI(1000))
    print(compute_PI_v2(1000))
```
