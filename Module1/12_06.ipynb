```python
def sum1(data):
    result = 0
    for value in data:
        if value % 2 == 0:
            result += value
    return result
data = [6, 5, 7, 1, 9, 2]
summation = sum1(data)
print(summation)
```

```python
def sum2(data):
    result = 0
    for i in range(len(data)):
        if i % 2 == 0:
            result += data[i]
    return result

data = [6, 5, 7, 1, 9, 2]
summation = sum2(data)
print(summation)
```

```python
def square(data):
    result = []
    for x in data:
        result.append(x * x)
    return result
def expression(data):
    return [x ** 2 for x in data]
result = [x ** 2 for x in [1, 2, 3, 4, 5, 6, 7, 8, 9] if x % 2 == 0]
print(result)
print(square([1, 2, 3]))
print(expression([1, 2, 3, 4]))
```

```python
def relu(x):
    relu = 0
    if x > 0:
        relu = x
    return relu
def reluForList(data):
    result = [relu(x) for x in data]
    return result

def min2(data):
    result = []
    while len(data) > 0:
        res = min(data)   # dùng min built-in
        result.append(res)
        data.remove(res)
    return result
print(min2([1, 5, 3, 6]))
```

```python
#Hadamard Product
def compute_Hadamard(a, b):
    rows = len(a)
    cols = len(a[0])
    result = [[None] * cols for _ in range(rows)]
    for row in range(rows):
        for col in range(cols):
            result[row][col] = a[row][col] * b[row][col]
    return result

a = [[3, 5, 7], [4, 9, 8]]
b = [[1, 6, 3], [0, 2, 9]]
print(compute_Hadamard(a, b)) 
```

```python
first_row = [1, 2, 3, 4]
first_col = [1,5, 9, 13]
rows = len(first_col)
cols = len(first_row)
arr = [[0] * cols for _ in range(rows)]
arr[0] = first_row
for c in range(cols):
    arr[c][0] = first_col[c]

for i in range(1, rows):
    for j in range(1, cols):
        arr[i][j] = arr[i - 1][j - 1] + arr[i - 1][j] + arr[i][j - 1]
for i in range(len(arr)):
    for j in range(len(arr[i])):
        print(arr[i][j], end = " ")
    print()
```

```python
fileName = 'data.txt'
def read_data(fileName):
    with open(fileName, 'r') as file:
        data = file.read()
        file.close()
    return data.split(',')

def fill_missing_with_average(data):
    missing_values = [i for i, x in enumerate(data) if len(x) == 0]
    non_missing_values = [float(x) for x in data if len(x) > 0]
    average = sum(non_missing_values) / len(non_missing_values)
    for index in missing_values:
        data[index] = average
    return data

data = read_data(fileName)
print(fill_missing_with_average(data))
def write_data(file_name, data):
    with open(file_name, 'w') as file:
        file.write(','.join(str(x) for x in data))
        file.close()

write_data('result.txt', data)
```
